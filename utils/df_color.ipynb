```python
def add_dict(dict1, dict2):
    dsum = {}
    d_k = list(dict1.keys()) + list(dict2.keys())  # sum keys list
    d_v = list(dict1.values()) + list(dict2.values())  # sum values list
    for i in range(len(d_k)):
        dsum[d_k[i]] = d_v[i]
    return dsum
```
