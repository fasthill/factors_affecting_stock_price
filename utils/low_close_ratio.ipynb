###  일별 최저가와 종가의 비율
#### 최저 대비 종가 상승비율: 종가가 +로 마감한다면, 최소 0.015% 밑에서 매수할 수 있는 기회가 <br> 회사별로 몇%인가를 알기 위함. 결과가 0.5라면 격일로 발생한다는 의미

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
code = {'005930' : ['삼성전자', 'sec'], '373220' : ['LG에너지솔루션', 'lgenergy'], 
        '000660' : ['SK하이닉스', 'skhinix'], '207940' : ['삼성바이오로직스', 'ssbio'],
        '006400' : ['삼성SDI', 'sdi'], '051910' : ['LG화학', 'lgchemical'],
        '005935' : ['삼성전자우', 'secpre'], '005380' : ['현대차', 'hyunmotor'],
        '035420' : ['NAVER', 'naver'], '000270' : ['기아','kia'],
        '035720' : ['카카오', 'kakao'], '005490' : ['POSCO홀딩스', 'poscoholding'],
        '105560' : ['KB금융', 'kbbank'], '028260' : ['삼성물산', 'sscnt'],
        '068270' : ['셀트리온', 'celltrion'], '012330' : ['현대모비스', 'mobis'],
        '055550' : ['신한지주', 'shgroup'], '066570' : ['LG전자', 'lgelec'],
        '003670' : ['포스코케미칼', 'poscochemical'], '096770' : ['SK이노베이션', 'skinnovation'],
        '033780' : ['KT&G', 'ktng']}
```

```python
code_n = ['sec', 'lgenergy', 'skhinix', 'ssbio', 'sdi', 'lgchemical','secpre', 'hyunmotor', 
          'naver', 'kia','kakao', 'poscoholding', 'kbbank',  'sscnt', 'celltrion', 'mobis',
          'shgroup', 'lgelec', 'poscochemical','skinnovation', 'ktng']
```

```python
df_dict = {}
dir_c = '../data/company_pkl/'
```

```python
df_base = pd.DataFrame(code_n)
df_base.columns=['company']
df_base = df_base.set_index('company')
```

```python
def find_ratio(cl_r, df_dict):
    for key, val in code.items():
        fname = f'{val[1]}_historical.pkl'
        f_name = dir_c + fname
        df = pd.read_pickle(f_name) 
        df_n = df[df['open'] != '']
        df_n = df_n.drop(labels='date', axis=1)
        df_n['cl'] = (df_n['close'] - df_n['low'])/df_n['low']
        df_c = len(df_n[df_n['cl'] >= cl_r])/len(df_n)
    #     print(f'**{val[1]}** ratio: {df_c:.2f}.')
        df_dict[val[1]] = df_c
        df_ratio = pd.DataFrame.from_dict(df_dict, orient='Index')
        df_ratio.columns = [f'ratio-{cl_r} 경우']
    return df_ratio
```

```python
for i in range(5, 21, 5):
    cl_r = i/1000
    df_ratio = find_ratio(cl_r, df_dict)
    df_base = df_base.merge(df_ratio, how='left', left_index=True, right_index=True)
    
df_base
```
