```python
from pandas_datareader import data as pdr
# from datetime import datetime
import yfinance as yf
yf.pdr_override()
```

```python
def get_ticker_data(ticker, startdate, enddate):
    ydata = pdr.get_data_yahoo(ticker, start=startdate, end=enddate)
    rdata = ydata.reset_index('Date')
    rdata['temp'] = rdata['Close'].shift(1)
    val_temp = (rdata['Close'] - rdata['temp'])/rdata['temp']*100
    rdata['close_cr'] = val_temp.map("{:.2f}".format)
    rdata = rdata[['Date','Open', 'High', 'Low', 'Close', 'close_cr', 'Volume']]
    rdata.columns = ['date', 'open', 'high', 'low', 'close', 'close_cr', 'vol']
    return rdata
```
