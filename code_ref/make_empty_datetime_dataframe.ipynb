### 날짜기준 empty dataframe 만들기

```python
import datetime, time
from datetime import date

import pandas as pd
import numpy as np
```

```python
start_date = '2020-01-01'  # 시작 일자
today = datetime.date.today()
today = today.strftime('%Y-%m-%d')
end_date = today  # 마감 일자
```

```python
column_name = ['a', 'b', 'c']
date_range_ts = pd.date_range(start=start_date, end=end_date)
df_out = pd.DataFrame(columns = column_name)
df_out.insert(0, 'date', date_range_ts)  # column 삽입
```

```python
def make_empty_datetime_df(start_date, end_date, column_name):
    # start_date =  '2021-01-01' str type.
    date_range_ts = pd.date_range(start=start_date, end=end_date)
    df_empty = pd.DataFrame(columns = column_name)
    df_empty.insert(0, 'date', date_range_ts)
    df_empty.replace(np.nan, '', inplace=True) # Nan을 빈 칸으로 대체
    return df_empty

start_date = '2020-01-01'
end_date = '2020-01-05'
column_name = ['a', 'b', 'c']
df_empty = make_empty_datetime_df(start_date, end_date, column_name )
```

```python
def make_datetime_series_df(df, start_date, end_date):
    # make dataframe with no empty date index with input df.
    # start_date =  '2021-01-01' str type.
    # df.index = datetime series
    # df.columns = ['a', 'b',,,,]
    date_range_ts = pd.date_range(start=start_date, end=end_date)
    df_empty = pd.DataFrame(columns = df.columns)
    df_empty.insert(0, 'date', date_range_ts)
    df_empty.replace(np.nan, '', inplace=True) # Nan을 빈 칸으로 대체
    df_empty.set_index('date', inplace=True)
    df_empty.update(df)
    return df_empty
```
