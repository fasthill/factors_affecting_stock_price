```python
def make_df_from_results(data):
    df = pd.DataFrame(data.split(","), columns=['var'])
    df['var1'] = df['var'].str.replace(":", "=")
    df['var2'] = df['var1'].str.split("=")
    df['param'] = df['var2'].str[0]
    df['value'] = df['var2'].str[1]
    df.drop_duplicates(subset=['param'], keep='last', inplace=True)
    df1 = df[['param', 'value']].set_index('param')
    return df1
```

```python
def convert_astype(x):
    try:
        np.isnan(x)
        return
    except:
        if x == 'None':
            return
        if '.' in x :
            return float(x)
        else :
            return int(x)
```

```python
def find_init_parameters(df):
    c1 = df.T['val_accuracy'].astype('float') >= 0.7  # 0.7 이상인 parameter 선정
    c2 = df.T['test_accuracy'].astype('float') >= 0.7
    c3 = df.T['precision'].astype('float') >= 0.7
    c_temp = df.T[c1 & c2 & c3]

    num_iterations = list(set(c_temp['num_iterations'].apply(lambda x : convert_astype(x) ))) # list로 저장
    num_iterations.sort()

    n_estimators = list(set(c_temp['n_estimators'].apply(lambda x : convert_astype(x) )))
    n_estimators.sort()

    learning_rate = list(set(c_temp['learning_rate'].apply(lambda x : convert_astype(x) )))
    learning_rate.sort()

    max_depth = list(set(c_temp['max_depth'].apply(lambda x : convert_astype(x) )))
    max_depth.sort()

    num_leaves = list(set(c_temp['num_leaves'].apply(lambda x : convert_astype(x) )))
    num_leaves.sort()

    subsample = list(set(c_temp['subsample'].apply(lambda x : convert_astype(x) )))
    subsample.sort()  

    bagging_fraction = list(set(c_temp['bagging_fraction'].apply(lambda x : convert_astype(x) )))
    bagging_fraction.sort()

    colsample_bytree = list(set(c_temp['colsample_bytree'].apply(lambda x : convert_astype(x) )))
    colsample_bytree.sort()

    feature_fraction = list(set(c_temp['feature_fraction'].apply(lambda x : convert_astype(x) )))
    feature_fraction.sort()
    
    max_delta_step = list(set(c_temp['max_delta_step'].apply(lambda x : convert_astype(x) )))
    max_delta_step.sort()
    
    parameters = {'num_iterations': num_iterations, 'n_estimators' : n_estimators, 'learning_rate' : learning_rate, 
                  'max_depth' : max_depth, 'num_leaves': num_leaves, 'subsample' : subsample, 
                  'bagging_fraction': bagging_fraction, 'colsample_bytree' : colsample_bytree, 
                  'feature_fraction' : feature_fraction, 'max_delta_step' : max_delta_step}
    return  parameters
#     return  num_iterations, n_estimators, learning_rate, max_depth, num_leaves, subsample, bagging_fraction, colsample_bytree, feature_fraction, max_delta_step
```

```python
def make_new_param(item, sel, int_flo): # int_flo: integer:0, float: 1
    item = sorted(item)
    interval = (max(item) - min(item))/len(item)
    
    # 소수점 이하 단위를 확인하기 위한 과정
    if (interval < 0.00001):
        cut_r = 6
    elif (interval < 0.0001):
        cut_r = 5
    elif (interval < 0.001):
        cut_r = 4
    elif (interval < 0.01):
        cut_r = 3
    elif (interval < 0.1):
        cut_r = 2
    elif (interval < 1):
        cut_r = 1
    else:
        cut_r = 0
    
    interval = round(((max(item) - min(item))/len(item)), cut_r)
    if interval < sel*0.1 :  # 원소값의 10%미만일 경우는 10%로 interval 설정
        interval = sel * 0.1
    print("--inerval--", interval)
    loc = item.index(sel) # 선정된 위치에 따라 다음 리스트 구성 기준 차이 
    # 가장 큰 쪽의 요소가 선정되었을 경우, sel, [sel+interval/2, sel+interval*2, sel+interval*3]
    # 가장 작은 쪽의 요소가 선정되었을 경우, sel-interval*3, sel-interval*2, sel-interval/2]
    #              if sel-interval*3 < 0 : [sel*0.25, sel*0.5, sel*0.75]
    # 가장 가운데 요소가 선정되었을 경우, 세분화 시킴 [ sel-int*0.5, sel, sel+int*0.5]
    print("*** selected value *** : ", sel)
    if loc == len(item)-1 : #가장 큰 쪽의 요소가 선정
        new_param = [sel-interval*0.5, sel+interval*0.5, sel+interval*2, sel+interval*3, sel+interval*4]
    elif loc == 0: # 가장 작은 쪽의 요소가 선정
        if (sel-interval*4) <= 0 :
            new_param = [sel*0.2, sel*0.4, sel*0.6, sel*0.8, sel*1.2]
        else:
            new_param = [sel-interval*4, sel-interval*3, sel-interval*2, sel-interval*0.5, sel+interval*0.5]
    else: # 선정된 요소가 중간에 있을 경우
        if (sel-interval) <= 0 :
            new_param = [sel*0.8, sel*0.9, sel*1.1, sel*1.2]
        else:
            new_param = [sel-interval, sel-interval*0.5, sel+interval*0.5, sel+interval]
    print("new_param", new_param)
            
    if int_flo == 0: # 0: integer, 1: flaot
        setp = np.int32(np.rint(new_param))
        new_list_nonzero = [i for i in list(set(setp)) if i != 0] # 0이 있으면 삭제
        if len(new_list_nonzero) <= 2 :
            new_list_nonzero = [1, 2, 3, 4]
        return sorted(new_list_nonzero)
        
    return list(np.round(new_param, 5))
```
