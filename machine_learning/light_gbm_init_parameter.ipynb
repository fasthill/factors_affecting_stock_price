```python
def find_init_parameters(df):
    c1 = df.T['val_accuracy'].astype('float') >= 0.7 
    c2 = df.T['test_accuracy'].astype('float') >= 0.7
    c3 = df.T['precision'].astype('float') >= 0.7
    c_temp = df.T[c1 & c2 & c3]

    num_iterations = list(set(c_temp['num_iterations'].astype('int'))) # list로 저장
    num_iterations.sort()

    n_estimators = list(set(c_temp['n_estimators'].apply(lambda x:  None  if x == 'None' else int(x) ))) 
    n_estimators.sort()

    learning_rate = list(set(c_temp['learning_rate'].apply(lambda x:  float(x)  if x != 'None' else None ))) 
    learning_rate.sort()

    max_depth = list(set(c_temp['max_depth'].apply(lambda x:  float(x)  if x != 'None' else None ))) 
    max_depth.sort()

    num_leaves = list(set(c_temp['num_leaves'].apply(lambda x:  int(x)  if x != 'None' else None ))) 
    num_leaves.sort()

    subsample = list(set(c_temp['subsample'].apply(lambda x:  float(x)  if x != 'None' else None ))) 
    subsample.sort()  

    bagging_fraction = list(set(c_temp['bagging_fraction'].apply(lambda x:  float(x)  if x != 'None' else None ))) 
    bagging_fraction.sort()

    colsample_bytree = list(set(c_temp['colsample_bytree'].apply(lambda x:  float(x)  if x != 'None' else None ))) 
    colsample_bytree.sort()

    feature_fraction = list(set(c_temp['feature_fraction'].apply(lambda x:  float(x)  if x != 'None' else None ))) 
    feature_fraction.sort()
    
    max_delta_step = list(set(c_temp['max_delta_step'].apply(lambda x:  float(x)  if x != 'None' else None ))) 
    max_delta_step.sort()
    
    parameters = {'num_iterations': num_iterations, 'n_estimators' : n_estimators, 'learning_rate' : learning_rate, 
                  'max_depth' : max_depth, 'num_leaves': num_leaves, 'subsample' : subsample, 
                  'bagging_fraction': bagging_fraction, 'colsample_bytree' : colsample_bytree, 
                  'feature_fraction' : feature_fraction, 'max_delta_step' : max_delta_step}
    return  parameters
#     return  num_iterations, n_estimators, learning_rate, max_depth, num_leaves, subsample, bagging_fraction, colsample_bytree, feature_fraction, max_delta_step
```
